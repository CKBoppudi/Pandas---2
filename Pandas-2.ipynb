# Outer, Inner, left and Right

```python
a = pd.DataFrame({"A" : [10,20] , "B" : [70,90] , "C" : [100,120]})
a
```

```python
b = pd.DataFrame({"A" : [10,20,30] , "B" : [70,90,100] , "C" : [100,120,145] , "D" : [145,675,232]}) 
b
```

```python
pd.concat([a,b])
```

# Note - By default concating always gives the Outer as it gives including all the values that are present in a and b.( Even though join = "outer" is not given.)

```python
a
```

```python
b
```

```python
pd.concat([a,b] , join = "inner")
```

# Note - From the when join = "inner" is given and as no axis is provided there are 3 columns common for a and b. So that is why the output is like above.

```python
pd.concat([a,b] , axis = 1 , join = "inner")
```

# Note - From the when join = "inner" is given axis = 1 is provided there are 2 rows common for a and b. So that is why the output is like above.

# Note - As we observe in the above while using "concat" and "join" we are getting the same columns having same values. To overcome this problem we have the "merge" function. 

```python
users = pd.DataFrame({"user id" : [1,2,3,4] , "name" : ["Dravid" , "Sachin" , "Ganguly" , "Yuvraj"]})
users
```

```python
titles = pd.DataFrame({"user id" : [1,2,3,1] , "title" : ["The wall" , "Master blaster" , "Dada" , "Mr Dependable"]})
titles
```

```python
pd.merge(users,titles)
```

# Find the name of the person whose title is "The wall".

```python
users.merge(titles , on = "user id")
```

```python
users.merge(titles)
```

# Note - Like seen above merge always gives the inner values(common) values present in the data frames.

# condition 1 for merge - When the column names are different but the values are same.

```python
titles.rename({"user id" : "id"} , axis = 1 , inplace = True) 
```

```python
titles
```

```python
users
```

```python
users.merge(titles , left_on = "user id" , right_on = "id")
```

# Note - By default the above gives the inner merge. But if we need to get the all possible values in the data frames outer should be mntioned. 

```python
users.merge(titles , left_on = "user id" , right_on = "id" , how = "outer")
```

# Left merge 

```python
users.merge(titles , left_on = "user id" , right_on = "id" , how = "left")
```

# right merge

```python
users.merge(titles , left_on = "user id" , right_on = "id" , how = "right")
```

# Note - By looking at the two above data sets, there are common columns in the two data sets which are "director_id" in the movies data set and "id" in the directors data set.

# Note - Now to know which movie is directed by which director we can merge the both data sets. So that the required director name for each movie is known. But before that we have to check if all the values in the "director id" column of movies are present in the "id" column of directors or not.

```python
movies["director_id"].isin(directors["id"])
```

```python
np.all(movies["director_id"].isin(directors["id"]))
```

# Note - The meaning of above 2 code snippets is first we checked how many values of "director id" are present in the "id" column. Next we checked if all the values are true or not using "np.all".

```python
movies.head(1)
```

```python
directors.head(1)
```

```python
movies.merge(directors , left_on = "director_id" , right_on = "id" , how = "left")
```

```python
data_new = movies.merge(directors , left_on = "director_id" , right_on = "id" , how = "inner")
```
